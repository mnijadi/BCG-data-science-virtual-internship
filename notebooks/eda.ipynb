Let's have closer look at the categories, and how they relate to churn.

```python
def contingency_plot(df, col1, col2, title, xlabel, ax):
    '''
    Plot a normalized stacked bar chart of the observed frequencies of two categorical variables.
    '''
    contingency_table = pd.crosstab(df[col1], df[col2])
    print(contingency_table, '\n\n')
    normalized_table = contingency_table.div(contingency_table.sum(axis=1), axis=0)
    normalized_table.plot(kind='bar', stacked=True, ax=ax)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Proportion')
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 7.5))

categories = ['has_gas', 'channel_sales', 'origin_up']
for i, category in enumerate(categories):
    contingency_plot(client_df, category, 'churn', f'Churn by {category}', category, ax=axes[i])

plt.suptitle('Churn by Categorical Variables', fontsize=20)
plt.tight_layout()
plt.savefig('../figures/contingency_plots.png')
plt.show()
```

- Starting with gas, we can't see if there is a significant difference between the two categories. We'll have to look at the numbers in a chi-squared test.
- Looking at the contingency table and stacker bar chart for channels and origins, we see that we have categories with few observations that aren't a churn, maybe we'll have to group them later on if we include these variables in our model.

Let's perform the chi-squared test.  
Note that the assumptions for doing such a test are satisfied:
- The observations are independent.
- Large sample size.
- Expected frequencies are definitely large enough.

```python
from scipy.stats import chi2_contingency

def chi2_test(df, col1, col2):
    contingency_table = pd.crosstab(df[col1], df[col2])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    print(f'Chi-square statistic: {chi2:.2f}')
    print(f'p-value: {p:.4f}')
```

```python
for category in categories:
    print(f'Churn by {category}')
    chi2_test(client_df, category, 'churn')
    print('\n')
```

Looking at the results of the test, the p-value is very small, the statistic is large enough, we can reject the null hypothesis and conclude that the two variables are dependent: `churn` depends on `has_gas`.  
Doing the same with `channel_sales` and `origin_up`, we get the same conclusion, eventhough we can't be so sure because these two have more categories with some having very few observations (less than 5).

- We notice almost the same distribution for the 4 variables of consumption.
- We see a very high density in lower values of the consumption.
- When looking at the statistics table, we notice that the 75% percentile is so low compared to the max value, we have lots of outliers that do not allow to see the true distribution of the data.

- Note also that for gas consumption, we have a high density for the value 0, we need to look at the distribution for values other than 0, since there is already another column that encodes the information of clients with 0 gas subscription.

```python
# filter out values above the 90% percentile
thresholds = [client_df[cons].quantile(0.90) for cons in consumption_columns]
figure, axes = plt.subplots(nrows=4, ncols=2, figsize=(8,12))

for i, cons in enumerate(consumption_columns):
    temp_serie = client_df[(client_df[cons] < thresholds[i]) & (client_df[cons] > 0)].loc[:, cons]
    temp_serie.hist(ax=axes[i//2, i%2], bins=50)
    axes[i//2, i%2].set_title(cons)
    temp_serie.plot(kind='box', ax=axes[i//2+2, i%2])
    axes[i//2+2, i%2].set_title(cons)

plt.suptitle('Distribution of consumption (>0 and <90% percentile)', fontsize=20)
plt.tight_layout()
plt.savefig('../figures/consumption_distributions.png')
plt.show()
```

Even after zooming in, we still have a high density of values in the lower values of consumption, and a large tail at the right with the high values.  
At the phase of feature engineering, we can easily mitigate the effect of outliers by using a log transformation.

Let's take a look at the dates, and see if there is any pattern.

```python
dates = ['date_activ', 'date_end', 'date_modif_prod', 'date_renewal']

fig, axes = plt.subplots(2, 2, figsize=(8, 8))

for i, date in enumerate(dates):
    client_df[date] = pd.to_datetime(client_df[date], format='%Y-%m-%d')
    temp_df = client_df.groupby(client_df[date].dt.year)['churn'].agg(['sum','count'])
    temp_df['churn_rate'] = temp_df['sum'] / temp_df['count']
    temp_df['churn_rate'].plot(kind='bar', ax=axes[i//2, i%2])
 
plt.suptitle('Churn Rate Over Time', fontsize=20)
plt.tight_layout()
plt.savefig('../figures/churn_rate_over_time.png')
plt.show()
```

- Looking at years af contracts' activation, we see that the churn rate was high and kept decreasing until 2007 at 2,5%, then it started increasing again all the way up to 15%
- The year of the end of the contract doesn't seem to have any impact on churn.
- The year of product's modification seems to follow the same pattern as the year of activation.
- Churn rate was the lowest for the most ancient years of contract's renewals.

Let's look at the historical data of the prices.

```python
# convert date columns to datetime
price_df['price_date'] = pd.to_datetime(price_df['price_date'], format='%Y-%m-%d')
```

Let's start by separating the 

```python
churned_clients = client_df.loc[client_df['churn']==1, 'id'].unique()
non_churned_clients = client_df.loc[client_df['churn']==0, 'id'].unique()
```

```python
churn_price_df = price_df.loc[price_df['id'].isin(churned_clients)]
non_churn_price_df = price_df.loc[price_df['id'].isin(non_churned_clients)]
```

```python
def plot_price_over_time(col, period, figname):
    fig = plt.figure(figsize=(10,8))
    gs = GridSpec(2, 2, figure=fig)

    ax1 = fig.add_subplot(gs[0, :])
    ax2 = fig.add_subplot(gs[1, 0])
    ax3 = fig.add_subplot(gs[1, 1])

    price_df.groupby('price_date')[col].mean().plot(ax=ax1, xlabel='')
    ax1.set_title(f'Average Price Over Time ({period} period)')

    churn_price_df.groupby('price_date')[col].mean().plot(label='Churned', ax=ax2, xlabel='')
    ax2.set_title('Churned')

    non_churn_price_df.groupby('price_date')[col].mean().plot(label='Non-churned', ax=ax3, xlabel='')
    ax3.set_title('Non-churned')

    plt.tight_layout()
    plt.savefig(f'../figures/{figname}.png')
    plt.show()
```

```python
plot_price_over_time('price_off_peak_var', '1st', 'price_1st_period_over_time')
```

We notice the same pattern for price evolution for both churns and non-churns:
- Prices were increasing from the beginning until the mid of 2015, then there is a sudden drop, from which they remain constant.

```python
plot_price_over_time('price_peak_var', '2nd', 'price_2nd_period_over_time')
```

```python
plot_price_over_time('price_off_peak_var', '3rd', 'price_3rd_period_over_time')
```

These evolutions are very similar for both churns and non-churns.
